## 1. Converting Columns to Datetime (`pd.to_datetime`)
- Use `pd.to_datetime()` to convert strings or numbers into datetime objects.
- Once converted, Pandas recognizes them as dates, enabling powerful operations.

```python
import pandas as pd

# 1. Create the dataset
data = {
    "OrderID": [1001, 1002, 1003, 1004, 1005],
    "OrderDate": ["2024-01-05", "2024/01/12", "15-01-2024", "2024-01-20", "20240125"],
    "Customer": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Amount": [250, 400, 150, 300, 500]
}

df = pd.DataFrame(data)
df['OrderDate'] = pd.to_datetime(df['OrderDate'],errors="coerce")
print(df)
```

```python
import pandas as pd

# 1. Create the dataset
data = {
    "OrderID": [1001, 1002, 1003, 1004, 1005],
    "OrderDate": ["2024-01-05", "2024/01/12", "15-01-2024", "2024-01-20", "20240125"],
    "Customer": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Amount": [250, 400, 150, 300, 500]
}

df = pd.DataFrame(data)
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='mixed',errors="coerce") #yaha bus change hai mixed ka order date same krne ki liya
print(df)
```

```python
pd.to_datetime(df['OrderDate'],
               errors = "coerce",
               format = 'mixed').dt.strfti,e('%d-%m-% y')
```

## 2. Setting DateTime Index & Resampling
- Set a datetime column as the index with `set_index()`.
- Resample time series using `resample('M').mean()` or other rules ('D', 'W', 'Y').
- Useful for aggregating data by time periods.

```python
print(df)
#step1 :- OrderDate set as Index.
df_idx = df.set_index('OrderDate')

#step2 : We will resample complete data according to months and we will get the total ammount of that month.
monthly_amount = df_idx.resample('M')['Amount'].sum()
(monthly_amount)

#step3 :
yearly_amount = df_idx.resample('Y')['Amount'].mean()
(yearly_amount)


#
year_average = df_idx.resample('Y')['Amount'].mean()
print(year_average)

#
week_average = df_idx.resample('W')['Amount'].mean()
print(week_average)
```
